```python
df_nova = df.drop('id',axis=1)
```

```python
#f)Salve uma versão da base, em outro arquivo, onde os atributos das colunas
#estejam normalizados entre 0 e 1 e onde os valores faltosos sejam
#substituídos pela média de sua coluna

normalized_df=(df_nova-df_nova.min())/(df_nova.max()-df_nova.min())
```

```python
normalized_df
```

```python
normalized_df.to_csv('mozilla4_nova.csv')
```
